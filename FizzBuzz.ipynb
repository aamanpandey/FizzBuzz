Print out a sequence of numbers 1,2,â€¦,99 on multiples of 5 print Fizz instead of the number. 
On multiples of 7 print Buzz instead of the number. 
At 35 or any multiple of both, print FizzBuzz

```python

x = 99

for i in range(1,x+1):
    if i%5==0 and i %7==0:
        print('FizzBuzz')
    if i%5==0:
        print('Fizz')
    elif i%7==0:
        print('Buzz')
    elif i%5!=0: print(i)
```
